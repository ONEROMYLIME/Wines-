```python
## This is imported data from a dataset in kaggel using over 130k wine reviews of wines all over the world. 
## Wine is reviewed using "points". 
#95-100 Classic: a great wine.
#90-94 Outstanding: a wine of superior character and style.
#85-89 Very good: a wine with special qualities.
#80-84 Good: a solid, well-made wine.
#75-79 Mediocre: a drinkable wine that may have minor flaws.
#50-74 Not recommended.
# I wanted to plot wine price  and "points"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv(r'C:\Users\moren\Downloads\wine-reviews\winemag-data-130k-v2.csv')


print(df)

```

```python
df.loc[:,['points', 'price']]



```

```python


x = ("price")
y = ('points')


plt.hist(x, normed=True, color='blue', bins=np.arange(-10, 10), alpha=.3) 
plt.hist(y, normed=True, color='green', bins=np.arange(-10, 10),alpha=.3)
plt.title('Wine Prices vs. Points')
plt.xlabel('values')

plt.show()
```

```python
## This histograph did not depict variety between price and wine points instead it shows a consistancy between the two
```

```python
plt.scatter(
    x=df['price'],
    y=df['points'],
    color='purple',
    marker='x', s=10
)
plt.scatter(
    x=df['price'],
    y=df['points'],
    color='green',
    marker='x', s=10
)
plt.show()
```

```python
#The scatter plot depicts the best array of both wine prices and "points". Showing that there are excellent wines with out breaking the bank!
```
